```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
plt.rcParams['figure.figsize'] = (8.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'
```

```python
img=plt.imread('Einstein.jpg')
```

```python
plt.imshow(img);
```

```python
gray_img = np.mean(img, axis=2)/255.0
plt.imshow(gray_img);
```

```python
def conv_single_step(img_slice,filter):
    output=np.sum(np.multiply(img_slice,filter))
    return output
```

```python
def forward_pass(img,W):
    height,width=img.shape
    Z=np.zeros((height-2,width-2))   
    for h in range(height-2):
        vert_start=h
        vert_end=vert_start+3
        for w in range(width-2):
            horiz_start=w
            horiz_end=horiz_start+3
            
            img_slice=img[vert_start:vert_end,horiz_start:horiz_end]
            Z[h,w]=conv_single_step(img_slice,W)
    return Z
```

```python
def visualize(original_img,filtered_img):
    fig, ax = plt.subplots(1, 2)
    ax[0].set_title('Original Image')
    ax[0].axis('off')
    ax[0].imshow(original_img)
    ax[1].set_title('Filtered Image')
    ax[1].axis('off')
    ax[1].imshow(filtered_img)
```

# Q.1

```python
#filter1 c=1
W1=np.array([0,0,1,
            0,0,0,
            0,0,0]).reshape(3,3)
W1
```

```python
filtered_img=forward_pass(gray_image,W1)
```

```python
visualize(gray_image,filtered_img)
```
